### ゼロ行列

$A \in M(m,n)$、すなわち A を m 行 n 列の行列とする。
A のすべての要素が 0 であるとき、A をゼロ行列といい、0 で表す。
ゼロ行列が m 行 n 列であることを示したいときは、$0_{m,n}$ と表す。

対角行列

正方行列

$$
A \in M(m,m)
$$

の対角線上にないすべての要素が $0$ であるとき、$A $を対角行列という。
すなわち、

$$
A=
\begin{pmatrix}
a_{11} & 0      & \cdots & 0 \\
0      & a_{22} & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0      & 0      & \cdots & a_{mm}
\end{pmatrix}
$$

である。あるいは、

$$
A=(a_{ij}), \qquad a_{ij}=0 \quad (i \ne j)
$$

と書くことができる。

単位行列

A が対角行列で、対角線上のすべての要素が 1 であるとき、A を単位行列といい、I で表す。
単位行列が m 行 m 列であることを示したいときは、$I_m$ と表す。

$$
I_m=
\begin{pmatrix}
1 & 0 & \cdots & 0 \\
0 & 1 & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & 1
\end{pmatrix}
$$
行列の和と積の性質
$$
A \in M(k,\ell), \qquad B \in M(m,n)
$$

とする。

A+B が定義できるのは、

$$
k=m,\qquad \ell=n
$$

のときのみである。

また、積 $AB$ が定義できるのは、

$$
\ell=m
$$

のときのみである。

行列の和、積が定義できるとき、次の性質が成り立つ。

$$
(1)\quad A0=0,\qquad 0A=0
$$
$$
(2)\quad (AB)C=A(BC)
$$

これを結合法則という。

$$
(3)\quad (A+B)C=AC+BC
$$
$$
(4)\quad A(B+C)=AB+AC
$$

(3)、(4) を分配法則という。

$$
(5)\quad AI_n=A,\qquad I_mA=A
$$

特に A が正方行列のときは、

$$
AI=IA=A
$$

と書ける。

$$
A \in M(m,m), \qquad B \in M(m,m)
$$

のとき、積に関して交換法則は一般には成り立たない。すなわち、

$$
AB=BA
$$

は一般には成り立たない。
