# スカラー（Scalar）

大きさ（数値）だけを持つ量です。
向き（方向）はありません。

例
温度（30℃）
質量（5kg）
時間（10秒）
エネルギー

👉 どれも「どの方向か」は関係なく、数値だけで意味が決まるのがポイントです。

## ベクトル（Vector）

大きさ + 向き（方向）を持つ量です。

例
速度（東に10 m/s）
力（右に5N）
位置（x, y 座標）
加速度

👉 「どっち向きか」が重要になるのが特徴です。

## 1.1.0. 幾何ベクトルと位置ベクトル

中学や高校では、平面直角座標における点を 
(𝑥,𝑦)
(x,y) のように表す。この表し方は、単に点の位置を示すだけでなく、「位置を表すベクトル」、すなわち位置ベクトルとして捉えることができる。

一方で、力や速度のように「大きさ」と「向き」の両方をもつ量もベクトルであり、これらは幾何ベクトルと呼ばれる。

幾何ベクトルは、矢印（有向線分）としてイメージすると理解しやすい。なお、位置ベクトルも幾何ベクトルと本質的に同じ性質をもつため、特に区別する必要はない。

ここでは、このベクトルの考え方をさらに発展させ、より高い次元の空間における図形の性質を調べるために拡張していく。

このようにして得られるベクトルの概念は、データ分析を行う際に不可欠な基礎的道具となる。

### 1.1.1 ベクトルの和とスカラー倍

実数全体の集合を 
$\mathbb{R} $
と表す。

本書では、縦に並べたベクトルを単に「ベクトル」と呼び、横に並べたものは「横ベクトル」と区別して呼ぶことにする。

いま、成分を 
𝑛
n 個もつ2つのベクトル

$$
\vec{a} =
\begin{pmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{pmatrix}, \quad
\vec{b} =
\begin{pmatrix}
b_1 \\
b_2 \\
\vdots \\
b_n
\end{pmatrix}
$$

を考える。このとき、それらの和 
$\vec{a}+\vec{b}$ は、対応する成分同士を足し合わせて

$$
\vec{a} + \vec{b} =
\begin{pmatrix}
a_1 + b_1 \\
a_2 + b_2 \\
\vdots \\
a_n + b_n
\end{pmatrix}
$$

と定義される。

また、実数（あるいは複素数）
λ とベクトル 
a の積（これをスカラー倍という）は、

$$
\lambda \mathbf{a} =
\begin{pmatrix}
\lambda a_1 \\
\lambda a_2 \\
\vdots \\
\lambda a_n
\end{pmatrix}
$$

と定義される。

特に、スカラーが0の場合には

$$
0\mathbf{a} =
\begin{pmatrix}
0 \\
0 \\
\vdots \\
0
\end{pmatrix}
$$

となる。このように、すべての成分が0であるベクトルを零ベクトルと呼び、
0 と表す。

零ベクトルは、任意のベクトル 
a に対して

$$
\mathbf{a} + \mathbf{0} = \mathbf{0} + \mathbf{a} = \mathbf{a}
$$

を満たす。

さらに、成分が 
n 個のベクトル全体の集合を 
$\mathbb{R}^n$　と表す。すなわち、

$$
\mathbb{R}^n =
\left\{
\begin{pmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{pmatrix}
\;\middle|\;
x_i \in \mathbb{R},\; 1 \le i \le n
\right\}
$$

である。

この 　$\mathbb{R}^n$　は、
n 次元実線形空間、あるいは 
n 次元ユークリッド空間と呼ばれる。

#### 補足

「$\mathbb{R}^n$とは、実数を 
n 個並べたベクトル全体の集合である」

$$
\mathbb{R}^n =
\left\{
\begin{pmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{pmatrix}
\;\middle|\;
x_i \in \mathbb{R},\; 1 \le i \le n
\right\}
$$

たとえば、$\mathbb{R}^2$　なら

$$
\begin{pmatrix}
x_1 \\
x_2 \\
\end{pmatrix}
$$

の形をしたもの全部

$\in$　←属するという意味

$x_i \in \mathbb{R}$

はつまり、
$x_i$ は実数全体 $\mathbb{R}$ に属する

R に属する」
→ 「$x_i$ は実数である」

真ん中の縦線 ∣

この縦線は

「ただし」
「そのような条件を満たす」

という意味です。

なので全体としては

$x_i$ が実数で、i が 1 から n まで動くような
そういう縦ベクトル全体

という意味になります。

### 1.1.2 基本ベクトル

$
\mathbb{R}^n
$において、特別な形をしたベクトルを基本ベクトルという。

基本ベクトルとは、ある1つの成分だけが1で、それ以外の成分がすべて0であるようなベクトルである。
第 $i$ 成分だけが1である基本ベクトルを $ \mathbf{e}_i $ と表す。
具体的には、次のようなベクトルである。

$$
\vec{e}_1 =
\begin{pmatrix}
1 \\
0 \\
\vdots \\
0
\end{pmatrix}, \quad
\vec{e}_2 =
\begin{pmatrix}
0 \\
1 \\
\vdots \\
0
\end{pmatrix}, \quad
\cdots \quad
\vec{e}_n =
\begin{pmatrix}
0 \\
0 \\
\vdots \\
1
\end{pmatrix}
$$

任意のベクトル $ \vec{a} $ は、

$$
\vec{a} =
\begin{pmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{pmatrix}
$$

のように表されるが、これは基本ベクトルを用いて次のように書き直すことができる。

$$
\mathbf{a}
= a_1 \mathbf{e}_1 + a_2 \vec{e}_2 + \cdots + a_i \vec{e}_i + \cdots + a_n \vec{e}_n
$$

さらに、これを総和記号を用いてまとめると、

$$
\mathbf{a}
= \sum_{i=1}^{n} a_i \vec{e}_i
$$

と表すことができる。

このように、任意のベクトルは基本ベクトルの線形結合として表される。
基本ベクトルは、ベクトル空間の構造を理解するうえで最も重要な基礎となる概念である

### 1.1.3 ベクトルの内積

$
\mathbb{R}^n
$におけるベクトルの基本ベクトルは次のように表される。

$$
\vec{e}_1 =
\begin{pmatrix}
1 \\
0 \\
\vdots \\
0
\end{pmatrix}, \quad
\vec{e}_2 =
\begin{pmatrix}
0 \\
1 \\
\vdots \\
0
\end{pmatrix}, \quad
\cdots \quad
\vec{e}_n =
\begin{pmatrix}
0 \\
0 \\
\vdots \\
1
\end{pmatrix}
$$

任意のベクトル $ \vec{a} $ は、

$$
\vec{a} =
\begin{pmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{pmatrix}
$$

と表され、基本ベクトルを用いて

$$
\vec{a} = \sum_{i=1}^{n} a_i \vec{e}_i
$$

と書くことができる。

#### 内積の定義

2つのベクトル $ \vec{a}, \vec{b} $ に対して、内積は次のように定義される。

$$
(\vec{a}, \vec{b}) = a_1 b_1 + a_2 b_2 + \cdots + a_n b_n
= \sum_{i=1}^{n} a_i b_i
$$

##### ベクトルの長さ（ノルム）

ベクトル $ \vec{a} $ の長さは次のように定義される。
$$
\|\vec{a}\| = \sqrt{a_1^2 + a_2^2 + \cdots + a_n^2}
$$

また、内積を用いると

$$
\|\vec{a}\|^2 = (\vec{a}, \vec{a})
$$

が成り立つ。
これは、ベクトルの長さが内積によって表されることを意味する重要な関係である。

###### 基本ベクトルの性質

基本ベクトルに対しては次が成り立つ。

$$
(\vec{e}_i, \vec{e}_j) = \delta_{i,j}
$$

ここで $ \delta_{i,j} $ はクロネッカーのデルタと呼ばれ、

$$
\delta_{i,j} =
\begin{cases}
1 & (i = j) \\
0 & (i \ne j)
\end{cases}
$$

である。

##### 内積の基本性質

任意のベクトル $ \vec{a}, \vec{b}, \vec{c} $ と実数 $ r $ に対して、次の性質が成り立つ。

$$
(1)\ (\vec{a}, \vec{a}) \ge 0
$$

（等号成立は $ \vec{a} = \mathbf{0} $ のときのみ）

$$
(2)\ (\vec{a}, \vec{b}) = (\vec{b}, \vec{a})
$$

$$
(3)\ (\vec{a} + \vec{b}, \vec{c}) = (\vec{a}, \vec{c}) + (\vec{b}, \vec{c})
$$

$$
(4)\ (\vec{a}, \vec{b} + \vec{c}) = (\vec{a}, \vec{b}) + (\vec{a}, \vec{c})
$$

$$
(5)\ (r\vec{a}, \vec{b}) = r(\vec{a}, \vec{b}) = (\vec{a}, r\vec{b})
$$

###### 展開公式

$$
((\vec{a} + \vec{b}), (\vec{a} + \vec{b}))
= \|\vec{a}\|^2 + \|\vec{b}\|^2 + 2(\vec{a}, \vec{b})
$$

$$
(\vec{a} - \vec{b}, \vec{a} - \vec{b})
= \|\vec{a}\|^2 + \|\vec{b}\|^2 - 2(\vec{a}, \vec{b})
$$

このように、内積はベクトル同士の関係（大きさや角度）を調べるための基本的な道具であり、線形代数やデータ分析において極めて重要な概念である。

### 1.1.4. 2つのベクトルの直交性

ここでは説明を簡単にするため、2次元のベクトルを考える。ただし、ここで述べる内積の性質は一般の $n$ 次元ベクトルにもそのまま拡張できる。

2つのベクトル $ \vec{a}, \vec{b} $ を

$$
\vec{a} =
\begin{pmatrix}
a_1 \\
a_2
\end{pmatrix}, \quad
\vec{b} =
\begin{pmatrix}
b_1 \\
b_2
\end{pmatrix}
$$

とする。

このとき、内積および各ベクトルの長さはそれぞれ次のように定義される。

$$
(\vec{a}, \vec{b}) = a_1 b_1 + a_2 b_2
$$

$$
\|\vec{a}\| = \sqrt{a_1^2 + a_2^2}, \quad
\|\vec{b}\| = \sqrt{b_1^2 + b_2^2}
$$

ここで、$\mathbf{a} \ne \mathbf{0}$、$\mathbf{b} \ne \mathbf{0}$ とする。
平面上で、点 $A(a_1, a_2)$、点 $B(b_1, b_2)$、原点 $O(0,0)$ を考え、角 $ \angle AOB = \theta $ とする。

三角形 $OAB$ に余弦定理を適用すると、

$$
|AB|^2 = |OA|^2 + |OB|^2 - 2|OA||OB|\cos \theta
$$

が成り立つ。

<details>
<summary>余弦定理（Cosine Rule）</summary>

三角形の3辺と角の関係を表す重要な公式です。
特に「直角三角形でない場合」でも使えるのがポイントです。
基本の式

三角形の3辺を $a, b, c$、角をそれぞれ $A, B, C$ とすると、

$$
c^2 = a^2 + b^2 - 2ab \cos C
$$

同様に、

$$
a^2 = b^2 + c^2 - 2bc \cos A
$$
$$
b^2 = a^2 + c^2 - 2ac \cos B
$$
どういう意味？

例えば

$$
c^2 = a^2 + b^2 - 2ab \cos C
$$

は

👉 2辺 $a, b$ とその間の角 $C$ がわかれば、残りの辺 $c$ が求まる

という意味です。

直角三角形との関係

もし角 $C = 90^\circ$ なら、

$$
\cos 90^\circ = 0
$$

なので、

$$
c^2 = a^2 + b^2
$$

<details>
<summary>cos（コサイン）とは？</summary>

角 $\theta$ に対して：

$$
\cos \theta = \frac{\text{隣の辺}}{\text{斜辺}}
$$

👉 ポイント

斜辺：一番長い辺
隣の辺：角 $\theta$ にくっついている辺

</details>

👉 これは ピタゴラスの定理になります。

ベクトルとのつながり（重要）

ベクトルではこの形になります👇

$$
\|\mathbf{a} - \mathbf{b}\|^2
=
\|\mathbf{a}\|^2 + \|\mathbf{b}\|^2
- 2(\mathbf{a}, \mathbf{b})
$$

そして

$$
(\mathbf{a}, \mathbf{b}) = \|\mathbf{a}\|\|\mathbf{b}\|\cos \theta
$$

なので、

👉 内積の式 = 余弦定理そのもの

です。

</details>

ここで、ベクトル $\overrightarrow{BA}$ は

$$
\vec{a} - \vec{b}
$$

であるから、上式はベクトルを用いて

$$
\|\vec{a} - \vec{b}\|^2
= \|\vec{a}\|^2 + \|\vec{b}\|^2 - 2\|\vec{a}\|\|\vec{b}\|\cos \theta
$$

と書き直せる。

左辺を成分で展開すると、

$$
\|\vec{a} - \vec{b}\|^2
= (a_1 - b_1)^2 + (a_2 - b_2)^2
$$

となる。

これを整理すると、

$$
\|\vec{a}\|^2 + \|\vec{b}\|^2 - \|\vec{a} - \vec{b}\|^2
= 2(a_1 b_1 + a_2 b_2)
$$

が得られる。

したがって、
$$
\cos \theta
= \frac{a_1 b_1 + a_2 b_2}{\|\mathbf{a}\|\|\mathbf{b}\|}
$$

となる。

これより、内積は次のようにも表される。

$$
(\mathbf{a}, \mathbf{b})
= a_1 b_1 + a_2 b_2
= \|\mathbf{a}\|\|\mathbf{b}\|\cos \theta
$$

この式から、2つのベクトルの内積が0であることと、2つのベクトルが直交している（すなわち $\theta = 90^\circ$）ことは同値であることがわかる。

### 1.1.5 シュワルツの不等式

2つのベクトル $ \vec{a}, \vec{b} $ の内積は、次のように表される。

$$
(\vec{a}, \vec{b}) = a_1 b_1 + a_2 b_2 = \|\vec{a}\|\|\vec{b}\|\cos \theta
$$

ここで $\theta$ は2つのベクトルのなす角である。

この式を変形すると、$\cos \theta$ は

$$
\cos \theta = \frac{(\vec{a}, \vec{b})}{\|\vec{a}\|\|\vec{b}\|}
$$

と表される。

三角関数の性質より、$|\cos \theta| \le 1$ が成り立つため、

$$
|(\vec{a}, \vec{b})| \le \|\vec{a}\|\|\vec{b}\|
$$

が得られる。これをシュワルツの不等式という。

成分で書くと、

$$
|a_1 b_1 + a_2 b_2|
\le
\sqrt{a_1^2 + a_2^2} \cdot \sqrt{b_1^2 + b_2^2}
$$

となる。

さらに、この不等式は一般の $n$ 次元の場合にも成り立つ。すなわち、

$$
\left|
\sum_{i=1}^{n} a_i b_i
\right|
\le
\sqrt{\sum_{i=1}^{n} a_i^2}
\cdot
\sqrt{\sum_{i=1}^{n} b_i^2}
$$

である。

また、$\vec{a} = \vec{0}$ または $\vec{b} = \vec{0}$ のときも、この不等式は成立する。

さらに、等号が成立するのは、2つのベクトルが平行な場合、すなわちある実数 $t \ne 0$ が存在して

$$
\vec{a} = t \vec{b}
$$

と表せるときである。

この不等式は、ベクトルの大きさと内積の関係を示す重要な基本結果であり、線形代数やデータ解析において広く用いられる。
