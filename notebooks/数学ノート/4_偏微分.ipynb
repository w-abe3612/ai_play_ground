偏微分

偏微分法の定義

ここでは，2個の独立変数の場合の微分法を説明する。3個以上の独立変数の場合にも，基本的には同様に扱うことができる。

2次元ユークリッド空間を次のように書く。

$$
\mathbb{R}^2=\{(x_1,x_2)\mid x_1,x_2\in\mathbb{R}\}
$$

また，その部分集合上で定義された2変数関数を考える。

$$
D\subset \mathbb{R}^2,\qquad y=f(x_1,x_2)\quad (y\in\mathbb{R})
$$

第2変数を固定して，第1変数だけの関数とみなし，第1変数について微分することを第1変数に関する偏微分といい，その結果を偏微分係数という。これを次のように書く。

$$
\frac{\partial f}{\partial x_1}=f_{x_1}
=\lim_{h\to 0}\frac{f(x_1+h,x_2)-f(x_1,x_2)}{h}
$$

同様に，第1変数を固定して，第2変数について微分したものを第2変数に関する偏微分係数といい，次のように書く。

$$
\frac{\partial f}{\partial x_2}=f_{x_2}
=\lim_{h\to 0}\frac{f(x_1,x_2+h)-f(x_1,x_2)}{h}
$$

1変数関数の微分係数の場合と同じく，偏微分可能であるためには，右の極限値と左の極限値が一致しなければならない。

$$
f_{x_1},\ f_{x_2}
$$

はそれぞれ2変数の関数とみることができるので，偏導関数という。

高階偏微分係数も1変数のときと同様に定義される。特に，2回微分して得られるものを2階偏導関数という。

$$
f_{x_1x_1}
=\frac{\partial}{\partial x_1}\left(\frac{\partial f}{\partial x_1}\right)
=\frac{\partial^2 f}{\partial x_1^2}
$$
$$
f_{x_2x_2}
=\frac{\partial}{\partial x_2}\left(\frac{\partial f}{\partial x_2}\right)
=\frac{\partial^2 f}{\partial x_2^2}
$$
$$
f_{x_1x_2}
=\frac{\partial}{\partial x_2}\left(\frac{\partial f}{\partial x_1}\right)
$$
$$
f_{x_2x_1}
=\frac{\partial}{\partial x_1}\left(\frac{\partial f}{\partial x_2}\right)
$$

<summary>偏微分とは何か</summary>

偏微分とは何か

ふつうの微分は、1つの変数だけで決まる関数を考えます。
たとえば

$$
y=x^2
$$

では、$x$ が変わると $y$ がどう変わるかを調べます。

一方で偏微分は、2つ以上の変数で決まる関数について、
そのうち1つだけを動かしたときに、値がどう変わるかを調べるものです。

たとえば

$$
y=f(x_1,x_2)
$$

という関数では、$x_1$ と $x_2$ の2つが入力です。
このとき、

$x_2$ をそのままにして、$x_1$ だけ動かす
$x_1$ をそのままにして、$x_2$ だけ動かす

という2通りの変化の見方があります。

この「ほかは固定して、1つだけに注目して微分する」のが偏微分です。

最初の文の意味

ここでは，2個の独立変数の場合の微分法を説明する。3個以上の独立変数の場合にも，基本的には同様に扱うことができる。

これは、

今は変数が2つある場合で説明する
でも考え方は3つ以上でも同じ

という意味です。

「独立変数」とは

独立変数とは、入力として自由に動かせる変数のことです。

ここでは $x_1, x_2$ が独立変数です。

たとえば

$$
y=f(x_1,x_2)
$$

なら、

$x_1$ を変える
$x_2$ を変える

ことで、結果の $y$ が変わります。

この式の意味
$$
\mathbb{R}^2=\{(x_1,x_2)\mid x_1,x_2\in\mathbb{R}\}
$$

これは少し記号が多いですが、意味は単純です。

全体の意味

「$\mathbb{R}^2$ とは、実数を2つ並べた組全部の集まりである」

という意味です。

記号を1つずつ説明
$\mathbb{R}$

$\mathbb{R}$ は 実数全体 を表します。

実数とは、たとえば

$1$
$-3$
$0$
$\frac{1}{2}$
$\sqrt{2}$
$3.14$

のような数です。

$\mathbb{R}^2$

$\mathbb{R}^2$ は、実数を2つ並べたもの全体です。

つまり、

$$
(x_1,x_2)
$$

の形をしたもの全部です。

たとえば

$(1,2)$
$(0,-5)$
$(3.1,\sqrt{2})$

などが入ります。

これはイメージとしては、平面上の点全体です。

$(x_1,x_2)$

これは「2つの数を順番に並べたもの」です。
座標だと思っても大丈夫です。

$x_1$ は1つ目の数
$x_2$ は2つ目の数

です。

ふつうの $(x,y)$ と似ていますが、ここでは変数名として $x_1, x_2$ を使っています。

${ \ }$

波かっこ ${ \ }$ は、集まりを表します。

たとえば「こういう条件を満たすもの全部を集めたもの」という意味です。

$\mid$

この縦棒 $\mid$ は、
「ただし」
「〜という条件で」
という意味です。

だから

$$
\{(x_1,x_2)\mid x_1,x_2\in\mathbb{R}\}
$$

は、

「$x_1,x_2$ が実数であるような $(x_1,x_2)$ 全体」

という意味になります。

$\in$

$\in$ は「〜に属する」「〜の仲間である」という意味です。

たとえば

$$
x_1\in\mathbb{R}
$$

は、

「$x_1$ は実数である」

という意味です。

次の式の意味
$$
D\subset \mathbb{R}^2,\qquad y=f(x_1,x_2)\quad (y\in\mathbb{R})
$$

これは、

$D$ は $\mathbb{R}^2$ の一部分
その $D$ の中の点 $(x_1,x_2)$ を入れると
実数 $y$ が決まる関数 $f$ を考える

という意味です。

これも記号を分けて説明
$D$

$D$ は、関数を考える範囲です。
定義域と呼ばれます。

たとえば平面全体ではなく、

円の内側だけ
$x_1>0$ の部分だけ

など、ある範囲だけで関数を考えることがあります。
その範囲を $D$ と書いています。

$\subset$

$\subset$ は「〜の部分集合」という意味です。

$$
D\subset \mathbb{R}^2
$$

は、

「$D$ は $\mathbb{R}^2$ の中に含まれている」

つまり

「$D$ は平面全体の中の一部分である」

という意味です。

$y=f(x_1,x_2)$

これは、

「$x_1, x_2$ を入れると、値 $y$ が決まる」

という意味です。

たとえば

$$
f(x_1,x_2)=x_1^2+x_2^2
$$

なら、

$(x_1,x_2)=(1,2)$ を入れると $1^2+2^2=5$
$(x_1,x_2)=(3,4)$ を入れると $3^2+4^2=25$

となります。

$(y\in\mathbb{R})$

これは、

「出てくる値 $y$ も実数である」

という意味です。

つまりこの関数は、平面上の点を入れると、答えとして1つの実数を返す関数です。

偏微分の説明文の意味

第2変数を固定して，第1変数だけの関数とみなし，第1変数について微分することを第1変数に関する偏微分という

これはとても大事です。

関数

$$
y=f(x_1,x_2)
$$

は本来、$x_1$ と $x_2$ の2つで決まります。

でも偏微分では、まず 片方を固定 します。

たとえば $x_2=3$ のまま動かさないことにすると、
$f(x_1,x_2)$ は実質的に

$$
f(x_1,3)
$$

となり、これは $x_1$ だけの関数として見られます。

そうしたうえで、$x_1$ について普通の微分をするのです。

これが $x_1$ に関する偏微分 です。

「固定する」とは

固定するとは、値をそのまま動かさないことです。

たとえば

$$
f(x_1,x_2)=x_1^2+3x_2
$$

で $x_2=5$ に固定すると、

$$
f(x_1,5)=x_1^2+15
$$

になります。

すると、これはもう $x_1$ だけの関数です。
だから普通に微分できます。

偏微分係数とは

偏微分係数とは、

「ある変数だけを少し変えたときに、関数の値がどれくらい変わるか」

を表すものです。

つまり、変化の鋭さや傾きのようなものです。

この式の意味
$$
\frac{\partial f}{\partial x_1}=f_{x_1}
=\lim_{h\to 0}\frac{f(x_1+h,x_2)-f(x_1,x_2)}{h}
$$

これは、$x_1$ に関する偏微分の定義です。

順番に見ます。

左側の記号
$\dfrac{\partial f}{\partial x_1}$

これは

「$f$ を $x_1$ で偏微分したもの」

という意味です。

記号の意味
$\partial$ は偏微分で使う特別な記号です
分母の $x_1$ は「何について微分するか」
分子の $f$ は「何を微分するか」

つまり

「$f$ を $x_1$ について偏微分する」

という読み方です。

「デル エフ デル エックスワン」と読むことが多いです。

$f_{x_1}$

これは上と同じ意味を、短く書いたものです。

$$
f_{x_1}=\frac{\partial f}{\partial x_1}
$$

です。

読み方は「エフ エックスワン」や「$x_1$ に関する偏導関数」です。

右側の極限の式
$\lim_{h\to 0}$

これは「$h$ を0に近づけるときの極限」です。

$h$ は、とても小さな変化量です。

つまり

$x_1$ をほんの少しだけ $h$ だけ増やしてみる
そのとき関数の値がどれくらい変わるかを見る
その変化量をどんどん小さくしていく

ということを表します。

$h$

$h$ は「ちょっとだけ増やす量」です。

$$
x_1 \to x_1+h
$$

という形で使われています。

つまり、$x_1$ をほんの少しだけ動かしています。

$f(x_1+h,x_2)$

これは、$x_1$ を $h$ だけ増やして、$x_2$ はそのままにしたときの関数の値です。

ここが偏微分らしいところです。

$x_1$ は変える
$x_2$ は変えない

となっています。

$f(x_1,x_2)$

これは元の場所での関数の値です。

$f(x_1+h,x_2)-f(x_1,x_2)$

これは、関数の値がどれだけ変わったかを表します。

つまり

「入力の $x_1$ を少しだけ動かしたら、出力がどれだけ変わったか」

です。

それを $h$ で割る意味
$$
\frac{f(x_1+h,x_2)-f(x_1,x_2)}{h}
$$

これは、

「$x_1$ を $h$ だけ動かしたとき、1あたりどれだけ $f$ が変わるか」

という意味です。

これは普通の微分で出てくる

$$
\frac{f(x+h)-f(x)}{h}
$$

と同じ考え方です。

全体として何をしているか

式全体は、

$x_2$ は固定する
$x_1$ を $h$ だけ増やす
そのときの関数の変化量を調べる
それを $h$ で割って変化の割合にする
$h$ を0に近づけて、ちょうどその点での変化のしかたを求める

という流れです。

これが $x_1$ に関する偏微分 です。

直感的なイメージ

2変数関数

$$
z=f(x_1,x_2)
$$

は、3次元の曲面だと思うと分かりやすいです。

その曲面の上で、

$x_2$ を固定して $x_1$ 方向にだけ動くときの傾き
$x_1$ を固定して $x_2$ 方向にだけ動くときの傾き

を調べるのが偏微分です。

つまり偏微分は、ある方向だけを見た傾きです。

具体例

たとえば

$$
f(x_1,x_2)=x_1^2+x_2
$$

を考えます。

$x_2$ を固定して $x_1$ で偏微分すると、$x_2$ は定数として扱うので、

$$
\frac{\partial f}{\partial x_1}=2x_1
$$

になります。

逆に、$x_1$ を固定して $x_2$ で偏微分すると、

$$
\frac{\partial f}{\partial x_2}=1
$$

になります。

つまり、

$x_1$ 方向の変化のしかた
$x_2$ 方向の変化のしかた

は別々に調べられる、ということです。

用語まとめ
偏微分

複数の変数を持つ関数で、1つの変数だけを動かして微分すること。

偏微分係数

ある変数だけを動かしたときの、その点での変化の割合。

2変数関数

2つの入力で値が決まる関数。
例：

$$
f(x_1,x_2)
$$
独立変数

入力として自由に動かす変数。ここでは $x_1, x_2$。

実数

ふつうの数全体。整数、分数、小数、無理数などを含む。

定義域

関数を考える入力の範囲。ここでは $D$。

極限

ある値にどんどん近づけたとき、どうなるかを見る考え方。
